### | DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환
___
- DataFrame은 파이썬 리스트, 딕셔너리 그리고 넘파이의 ndarray 등 다양한 데이터로부터 생성될 수 있다.
- 또한 DataFrame은 반대로 파이썬 리스트, 딕셔너리 그리고 넘파이의 ndarray 등으로 변환될 수 있다.



#### 1. 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기
 - DataFrame은 리스트, 넘파이 ndarray와 다르게 Column명을 가지고 있다.
 - 이 차이가 상대적으로 데이터 핸들링을 편하게 만든다.
 - 일반적으로 DataFrame으로 변환할 때, Column명을 지정해준다.(다만, 지정하지 않으면 자동으로 Column명을 할당해준다.)
 ```python
 pd.DataFrame(data, columns = 'Column명 리스트')
 ```
 - DataFrame은 기본적으로 행과 열을 가지는 2차원 데이터이기 때문에, 2차원 이하의 데이터들만 DataFrame으로 변환할 수 있다.

````python
# 1차원 데이터를 기반으로 DataFrame 생성
col_name1 = ['col1'] # 1차원 데이터니까, Column명이 1개만 필요
list1 = [1, 2, 3]
array1 = np.array(list1)

print('array1 shape : ', array1.shape)

# list를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns = col_name1)
print('1차원 리스트로 만든 DataFrame :\n', df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns = col_name1)
print('1차원 리스트로 만든 DataFrame :\n', df_array1)
````

````python
# 2차원 데이터를 기반으로 DataFrame 생성
col_name2 = ['col1','col2', 'col3'] # 2x3 형태의 데이터니까, Column명 3개 필요

# 2X3 형태의 list와 ndarray 생성
list2 = [[1, 2, 3],
         [11, 12,13]]
array2 = np.array(list2)

# 2X3 형태의 list를 이용해 DataFrame 생성
df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame :\n', df_list2)

# 2X3 형태의 넘파이 ndarray를 이용해 DataFrame 생성

df_array2 = pd.DataFrame(array2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame :\n', df_array2)
````

````python
# 딕셔너리를 이용해 DataFrame 생성
# Key는 문자열 Column명으로, Value는 리스트 형 혹은 ndarray형 column 데이터로 mapping
dic = {'col1' : [1,11],
       'co12' : [2,22],
       'col3' : [3,33]}
df_dict = pd.DataFrame(dic)
print('딕셔너리로 만든 DataFrame :\n', df_dict)
````

#### 2. DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기
- 많은 머신러닝 패키지가 기본 데이터 형으로 넘파이 ndarray를 사용
- 따라서 데이터 핸들링은 DataFrame을 이용하더라도 머신러닝 패키지를 활용하기 위해 다시 넘파이 ndarray로 변환해야 하는 경우가 발생한다.
- DataFrame 객체의 `values`를 이용해 넘파이의 ndarray로 변환할 수 있다.
- `values`를 이용해 ndarray로 변환한 다음, `tolist()`를 호출하면 list로 변환할 수 있다.
- DataFrame 객체의 `to_dict()` 메서드를 이용해 딕셔너리로 변환할 수 있다.


````python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values의 type : ', type(array3), 'df_dict.values shape : ', array3.shape)
print(array3)
````

````python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist() # ndarray로 변환한 다음 tolist()를 호출해 list로 변환
print('df_dict.values.tolist()의 type : ', type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list') # 딕셔너리 값이 리스트형으로 반환됨
print('\n df_dict.to_dict() 의 type : ', type(dict3))
print(dict3)
````

### | DataFrame 데이터 삭제
___
- `drop()` 메서드를 이용해 DataFrame 내의 데이터를 삭제할 수 있다.
```python
DataFrame.drop(labels = None, axis = 0, index = None, columns = None, level = None, inplace = False, errors = 'raise')
```
  + **axis** : axis 값에 따라서 특정 Column(axis 1) 혹은 row(axis 0)를 drop한다.
  + **labels** : 원하는 Column명을 입력하고 axis = 1을 입력하면 지정된 Column을 drop한다. 여러 개의 Column을 삭제하고 싶을 경우 list 형태로 입력하면 된다.
  + **inplace** : default 값은 False로, `inplace = False` 이면 원본 DataFrame의 데이터는 삭제하지 않고, 삭제된 결과를 새 DataFrame으로 반환한다.
- 아래 예제 외에도 원본 DataFrame에서 드롭된 DataFrame을 다시 원본 DataFrame 객체 변수로 할당하면 원본 DataFrame에서 데이터를 드롭하는 것과 같은 결과가 나온다. (단, 기존 원본 DataFrame 객체 변수는 메모리에서 추후 제거된다.)
```python
titanic_df = titanic_df.drop('Age_0', axis = 1, inplace = False)
```


````python
# 'Age_0' Column 삭제하기
titanic_drop_df = titanic_df.drop('Age_0', axis = 1)
titanic_drop_df.head(3)
````

````python
titanic_df.head(3) # Inplace = False라서 원본 DataFrame에서 'Age_0' Column이 삭제되지 않았다.
````

### | Index 객체
___
- 판다스의 Index 객체는 RDBMS의 PK(Primary Key)와 유사하게 DataFrame, Series의 레코드를 고유하게 식별하는 객체이다.
- `DataFrame.index` 나 `Series.index`를 이용해 DataFrame과 Series에서 Index 객체만 추출할 수 있다.

````python
# 원본 파일 다시 로딩
titanic_df = pd.read_csv('train.csv')

# Index 객체 추출
indexes = titanic_df.index
print(type(indexes))

# Index 객체를 실제 array 값으로 변환
print('Index 객체 array 값 :\n', indexes.values) # (Review) values : DataFrame의 객체를 ndarray로 변환
````

- Index 객체는 식별성 데이터를 1차원 array 형태로 가지고 있다.
- 또한 ndarray와 유사하게 단일 값 반환 및 슬라이싱이 가능하다.


````python
print(type(indexes.values))
print(indexes.values.shape)
print(indexes[:5].values)
print(indexes.values[:5])
print(indexes[6])
````

- 한 번 만들어진 DataFrame과 Series의 Index 객체는 함부로 변경할 수 없다.

````python
indexes[0] = 5
````

- Series 객체는 Index 객체를 포함하지만, Series 객체에 연산 함수를 적용할 때 Index는 연산에서 제외된다. 즉, Index는 오직 식별용으로만 사용된다.

````python
series_fair = titanic_df['Fare']
print('Fair Series max 값 : ', series_fair.max())
print('Fair Series sum 값 : ', series_fair.sum())
print('sum() Fair Series : ', sum(series_fair))
print('Fair Series + 3 :\n', (series_fair+3).head(3))
````

- **reset_index()** 메서드 : DataFrame과 Series에 Index를 새롭게 연속 숫자 형으로 할당해준다. 그리고 기존 인덱스는 'index'라는 새로운 Column명으로 추가된다.
  + Index가 연속된 int 숫자형 데이터가 아닐 경우, 이를 연속 int 숫자형 데이터로 만들 때 주로 사용한다.
  + ex ) Column 'Pclass'에서 Series의 `value_counts()`를 실행하면, 'Pclass' 고유 값이 Index 역할을 했는데, `reset_index()` 이용해 연속 숫자형 Index를 할당해줄 수 있다.
  + Series에 `reset_index()`를 적용하면 DataFrame이 반환되니 유의해야 한다.
  + `drop = True`로 설정하면 기존 Index는 새로운 Column으로 추가되지 않고 삭제(drop)되고, 그대로 Series로 유지된다.

````python
titanic_reset_df = titanic_df.reset_index(inplace = False)
titanic_reset_df.head(3)
````

````python
print('#### before reset_index ####')
values_counts = titanic_df['Pclass'].value_counts()
print(values_counts)
print('values_counts 객체 변수 타입 : ', type(values_counts))
new_value_counts = values_counts.reset_index(inplace = False)

print('#### after reset_index  ####')
print(new_value_counts)
print('new_values=_counts 객체 변수 타입 : ', type(new_value_counts))
````

### | 데이터 셀렉션 및 필터링

- 지금은 DataFrame 뒤에 있는 []는 Column만 지정할 수 있는 **Column 지정 연산자**로 이해하는 것이 좋다.

####2. DataFrame의 ix[ ] 연산자
- **ix[ ]** : 행 위치로 인덱스값을, 열 위치로 Column명을 입력해 원하는 위치의 데이터를 추출할 수 있다.
  + 판다스 개발 초기에 넘파이 ndarray의 [ ] 연산자와 유사한 기능을 DataFrame에 제공하기 위한 연산자
  + 칼럼 명칭 기반 인덱싱(Column label indexing)
  ```python
  ix[0, 'Pclass']
  ```
  + 칼럼 위치 기반 인덱싱(Column position indexing)
  ```python
  ix[0,2]
  ```
  + 두 가지 방식을 모두 제공하게 되면서 코드에 혼란을 주거나 가독성이 떨어지면서 `ix[ ]`는 </U> 현재 판다스에서 사라졌다 </U>. 그리고 이를 대신할 새로운 연산자가 만들어졌다.
- **loc[ ]** : 칼럼 명칭 기반 인덱싱 연산자
- **iloc[ ]** : 칼럼 위치 기반 인덱싱 연산자



````python
# 교재 속 예제들은 ix[ ]를 사용하고 있지만, 현재 기능이 삭제되어 loc[]와 iloc[]로 대체
print('칼럼 위치 기반 인덱싱 데이터 추출 : ', titanic_df.iloc[0,2])
print('칼럼 명 기반 인덱싱 데이터 추출 : ', titanic_df.loc[0,'Pclass'])
````

````python
# 예시 데이터 프레임 생성
data = {'Name' : ['Chulmin', 'Eunkyung', 'Jinwoong', 'Soobeom'],
        'Year' : [ 2011, 2016, 2015, 2015],
        'Gender' : ['Male', 'Female', 'Male', 'Male']
        } # dictionary 생성
data_df = pd.DataFrame(data, index = ['one', 'two', 'three', 'four'])
print(data_df)

# data_df를 reset_index()로 새로운 숫자형 인덱스 생성
data_df_reset = data_df.reset_index()
data_df_reset = data_df_reset.rename(columns = {'index':'old_index'}) # 기존의 인덱스값을 'old_index' 열로 옮김 
data_df_reset.index = data_df_reset.index+1 # 인덱스가 1부터 시작하도록 설정
print(data_df_reset)
````

  + index가 int형인 경우, `[ ]` 안에 행 위치 값으로 정수를 입력하면 DataFrame의 index 값을 호출한다.
  + DataFrame의 Index값은 명칭 기반 인덱싱으로 간주

##### (5) DataFrame loc[ ] 연산자
- `loc[ ]` : 명칭 기반으로 데이터를 추출한다.
- 행 위치에는 DataFrame index 값을, 열 위치에는 Column 명을 입력한다.
- DataFrame index 값은 정수여도 명칭 기반 인덱싱으로 간주하기 때문에, 이 경우, 행 위치에 정수가 들어가도 된다.
- loc[ ]와 슬라이싱 : `[start : end]`로 입력하면 일반적으로 end-1까지 포함하는 것이 아니라, end 값까지 포함한다.
  + 명칭 기반 인덱싱의 경우, 명칭이 숫자 형이 아닐 수 있기 때문에 -1을 할 수 없기 때문에 위와 같은 결과가 나온다.


````python
data_df.loc['one','Name']
````

````python
data_df_reset.loc[1, 'Name']
````

````python
# 다음 코드는 오류가 발생한다.
data_df_reset.loc[0, 'Name'] # index값이 0인 행이 없기 때문
````

````python
# 슬라이싱을 적용할 때, 명칭 기반 인덱싱과 위치 기반 인덱싱의 차이
print(data_df, '\n')
print('위치 기반 iloc slicing\n', data_df.iloc[0:1, 0], '\n')
print('명칭 기반 loc slicing\n', data_df.loc['one':'two','Name'])
````

````python
print(data_df_reset.loc[1:2, 'Name'])
````

####6. 불린 인덱싱
- `iloc[ ]`나 `loc[ ]`는 조건에 해당하는 값이 무엇인지 계산하여 를 사용해야 하지만, 불린 인덱싱은 조건을 입력하면 자동으로 원하는 값을 필터링하기 때문에 많이 사용한다.
- 불린 인덱싱은 `[ ]`, `ix[ ]`, `loc[ ]`에서 지원한다.
- `iloc[ ]`는 앞서 언급한 것처럼 정수형이 아닌 Boolean 형은 지원하지 않아 불린 인덱싱이 불가하다.
- 여러 개의 조건을 두는 것도 가능하다.
  1. and 조건 : `&`
  2. or 조건 : `|`
  3. Not 조건 : `~`

````python
titanic_df = pd.read_csv('train.csv')
titanic_boolean = titanic_df[titanic_df['Age'] > 60]
print(type(titanic_boolean)) # DataFrame
titanic_boolean.head()
````

````python
titanic_df[titanic_df['Age'] > 60][['Name', 'Age']].head(3)
````

````python
# loc[ ]와 불린 인덱싱
titanic_df.loc[titanic_df['Age'] > 60, ['Name','Age']].head(3)
````

````python
# 여러 개의 조건으로 불린 인덱싱
titanic_df.loc[ (titanic_df['Age'] > 60) & (titanic_df['Pclass'] == 1) & (titanic_df['Sex'] == 'female') ]
````

````python
# 개별 조건을 변수에 할당
cond1 = titanic_df['Age'] > 60
cond2 = titanic_df['Pclass'] == 1
cond3 = titanic_df['Sex'] == 'female'
titanic_df[ cond1 & cond2 & cond3]
````

### | 정렬, Aggregation 함수, GroupBy 적용

####1. DataFrame, Series의 정렬 - sort_values()
- `sort_values()` 메서드 : DataFrame과 Series를 정렬
- RDBMS SQL의 `order by` 키워드와 매우 유사하다.
```python
DataFrame.sort_values(by = ['Column명'], ascending = False, inplace = False)
```
  1. **by** : 특정 Column을 입력하면 해당 Column을 정렬
  2. **ascending** : True이면 오름차순으로 정렬하고, False이면 내림차순으로 정렬. 기본은 True
  3. **inplace** : False로 설정하면 메서드를 호출한 DataFrame은 그대로 유지되고, 정렬된 DataFrame을 결과로 반환. 기본은 False

````python
titanic_sorted = titanic_df.sort_values(by = ['Name'])
titanic_sorted.head(3)
````

````python
# 여러 개의 Column을 정렬
titanic_sorted = titanic_df.sort_values(by = ['Pclass','Name'], ascending = False)
titanic_sorted.head(3)
````

- 다만 DataFrame의 경우, DataFrame에서 바로 aggregation 함수를 호출할 경우, 모든 Column에 해당 aggregation 함수가 적용된다는 차이가 있다.

####3. groupby() 적용
- DataFrame의 `groupby()`는 RDBMS SQL의 groupby 키워드와 유사하지만, 다른 점이 있어 주의해야 한다.
- SQL과 판다스 모두 groupby를 분석 작업에 많이 활용한다.
```python
DataFrame.groupby(by = 'Column명')
```
- `by`에 Column을 입력하면 그 Column을 기준으로 groupby된다.
- DataFrame에서 groupby()를 호출하면 **DataFrameGroupBy**라는 형태의 DataFrame을 반환한다.

````python
titanic_groupby = titanic_df.groupby(by = 'Pclass')
print(type(titanic_groupby))
````

- DataFrame은 `groupby()`를 호출해 반환된 결과에 aggregation 함수를 호출하면 groupby() 대상 Column을 제외한 모든 Column에 그 함수가 적용된다.
- 이와 다르게 SQL에서 동일한 작업을 수행하려면 대상 Column을 모두 Select 절에 나열해야 한다.

````python
titanic_groupby = titanic_df.groupby(by = 'Pclass').count()
titanic_groupby 
````

- DataFrame의 groupby()에 특정 Column만 aggregation 함수를 적용하려면 grouby()로 반환된 DataFrameGroupBy 객체에 해당 Column을 필터링한 뒤, 함수를 적용해야 한다.

````python
titanic_groupby = titanic_df.groupby(by = 'Pclass')[['PassengerId', 'Survived']].count()
titanic_groupby
````

- DataFrame의 groupby()에 서로 다른 여러 개의 aggregation 함수를 적용하려면, 함수명을 DataFrameGroupBy 객체의 `agg()` 내에 인자로 입력해야 한다.

````python
titanic_df.groupby(by = 'Pclass')['Age'].agg([max, min])
````

- 여러 개의 Column에 대해 서로 다른 aggregation 함수를 적용하려면, DataFrame은 SQL에 비해 좀 더 복잡한 처리가 필요하다
  1. 적용할 Column을 key로 적용할 함수를 value로 하는 딕셔너리를 선택한다.
  2. agg()의 인자로 위에서 만든 딕셔너리를 입력한다.

````python
agg_format = {'Age' : 'max', 'SibSp' : 'sum', 'Fare' : 'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
````

### | 결손 데이터 처리하기
___
- 판다스는 결손 데이터(Missing Data)를 처리하는 편리한 API를 제공한다.
- 결손 데이터 : Colum에 값이 없는, 즉 NULL인 경우. 넘파이의 NaN으로 표시한다.
- 기본적으로 머신러닝 알고리즘은 이 NaN 값을 처리하지 않으므로, 이 값을 다른 값으로 변경해줘야 한다.
- 또한 NaN 값은 평균, 총합 등의 함수 연산에서 제외된다.
- `isna()` : NaN 여부를 확인하는 API
- `fillna()` : NaN 값을 다른 값으로 대체하는 API

####1. isna()로 결손 데이터 여부 확인
- **isna()** : 데이터가 NaN인지 아닌지를 알려준다.
- DataFrame에 적용하면 Column의 값이 NaN인지 아닌지를 True나 False로 모두 알려준다.
- 결손 데이터의 개수는 isna() 결과에 `sum()` 함수를 추가해 구할 수 있다. 함수를 호출하면, True는 숫자 1로, False는 0으로 변환되기 때문이다.

````python
titanic_df.isna().head(3)
````

````python
# 결손 데이터의 개수 구하기
titanic_df.isna().sum()
````

####2. fillna()로 결손 데이터 대체하기
- **fillna()***를 이용하면 결손 데이터를 편리하게 다른 값으로 대체할 수 있다.
- fillna()를 적용할 때, 반환 값을 원래의 DataFrame으로 다시 받거나, `inplace = True`로 설정해야 실제 DataFrame 값이 변경된다.

````python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
````

````python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum()
````

### | apply lambda 식으로 데이터 가공
___
- 판다스는 **apply 함수**에 **lambda 식**을 결합해 DataFrame이나 Series의 데이터를 레코드별로 가공하는 기능을 제공한다.
- Column에 일괄적으로 데이터 가공을 하는 것이 더 빠르거나, 복잡한 데이터 가공이 필요할 경우, **apply lambda**를 사용한다.
- **lambda 식** : 파이썬에서 함수형 프로그래밍(functional programming)을 지원하기 위해 만들었다.
  + 함수명과 Input을 선언하는 등 복잡한 과정을 거치지 않고, 함수의 선언과 처리를 한 줄의 식으로 쉽게 변환할 수 있다.
  ```python
  lambda 'input' : 'return값'
  ``` 
  + `map()` 함수를 이용해 여러 개의 값을 Input 값으로 받을 수 있다.

````python
# 입력 값의 제곱 값을 구해서 반환하는 get_square(a) 함수
def get_square(a):
  return a**2

print('3의 제곱은 : ', get_square(3))
````

````python
# get_square(a)함수를 lambda 식으로 변환
lambda_square = lambda x : x**2
print('3의 제곱은 : ', lambda_square(3))
````

````python
# 여러 개의 값을 입력
a = [1, 2, 3]
squares = map(lambda x : x**2, a)
list(squares)
````

- 판다스 DataFrame의 lambda 식 : 파이썬의 lambda 식을 그대로 적용한 것
- lambda 식은 if else를 지원하는데, 이때, if 식보다 return값을 먼저 작성해야 한다.
  + lambda 식의 ':' 기호 오른쪽에 반환값이 있어야 하기 때문
  + else if는 지원하지 않는다
  + else if를 이용하기 위해서는 else 절을 ()로 내포해 () 내에서 다시 if else를 적용해야 한다.
  + 다만 else if가 많이 나오거나 switch case문의 경우에는 별도의 함수를 만드는 것이 더 좋다.

````python
# apply lambda 식으로 DataFrame 가공하기
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head(3)
````

````python
# lambda 식에서 if else 절 활용
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
````

````python
# lambda 식에서 else if 구현
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
````

````python
# 나이에 따라 세분화하여 분류하는 함수
def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young Adult'
  elif age <= 60: cat = 'Adult'
  else : cat = 'Elderly'
  return cat

# lambda 식에 위에서 생성한 함수를 반환값으로 지정
# get_category(X)는 입력값으로 'Age' Column값을 받아서 해당하는 cat을 반환하도록 설정
titanic_df['Age_cat'] =  titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
````
