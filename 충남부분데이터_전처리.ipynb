```python
def city(add):
  splitted = add.split()
  if len(splitted) > 1:
    if '천안' in splitted[0]: 
      return '천안시'
    else:
      return splitted[1]
  else:
    return "unknown"

cn['city'] = cn.apply(lambda x: city(x['HPPN_PNU_ADDR']), axis=1)
cn.city.value_counts()
```

```python
def town(add):
  splitted = add.split()
  if len(splitted) > 1:
    if '천안' in splitted[0]: 
      return splitted[2]
    elif '천안' in splitted[1]:
      if '동남' in splitted[2]:
        return splitted[3]
      elif '서북' in splitted[2]:
        return splitted[3]
      else:
        return splitted[2]
    else:
      return splitted[2]
  else:
    return "unknown"

cn['town'] = cn.apply(lambda x: town(x['HPPN_PNU_ADDR']), axis=1)
```

```python
# town안의 괄호 제거
import re

def bracket(town):
  result = re.sub(r'\([^)]*\)', '', town)
  return result

cn['town2'] = cn.apply(lambda x: bracket(x['town']), axis=1)
cn.head()
```

```python
def above_three(town):
  return len(town)>3
cn['is_town'] = cn.apply(lambda x: above_three(x['town2']), axis=1)
cn['is_town'].value_counts()
```

```python
def count_word(town):
  return len(town)
cn['word_town'] = cn.apply(lambda x: count_word(x['town2']), axis=1)
cn['word_town'].value_counts()
```
