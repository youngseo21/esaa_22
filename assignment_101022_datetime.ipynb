# | **datetime 오브젝트**
___
- datetime 라이브러리 : 날짜와 시간을 처리하는 등의 다양한 기능을 제공하는 파이썬 라이브러리
- 주로 다루는 datetime 라이브러리의 오브젝트
  1. `date` 오브젝트 : 날짜를 처리
  2. `time` 오브젝트 : 시간을 처리
  3. `dattetime` 오브젝트 : 날짜와 시간을 모두 처리

#### [실습] datetime 오브젝트 사용하기
___

```python
# import datetime 라이브러리
from datetime import datetime
```

##### 1. 현재 시간을 출력하는 메서드 : `now()`, `today()`

```python
# 현재 시간을 출력하는 메서드 1 : now
now1 = datetime.now()
print(now1)
```

```python
# 현재 시간을 출력하는 메서드 2 : today 
now2 = datetime.today( ) 
print(now2)
```

##### 2. 시간을 직접 입력하여 datetime 오브젝트 생성하기

```python
t1 = datetime.now() # now 메서드는 밀리초 단위까지
t2 = datetime(1970, 1, 1) 
t3 = datetime(1970, 12, 12, 13, 24, 34)

print(t1) 
print(t2)
print(t3)
```

##### 3. datetime 오브젝트로 시간 계산하기

```python
t1 = datetime.now()
t2 = datetime(1970, 1, 1) 
t3 = datetime(1970, 12, 12, 13, 24, 34)
```

```python
diff1 = t1 - t2
print(diff1)
print(type(diff1))
```

```python
diff2 = t2 - t1
print(diff2)
print(type(diff2))
```

### | **시간 범위와 인덱스**
___
- 특정 일에 누락된 데이터가 있을 수 있다.
- 임의로 시간 범위를 생성하여 인덱스로 지정하면 누락된 데이터도 포함하여 데이터를 살펴볼 수 있다.

#### [실습] 시간 범위 생성해 인덱스로 지정하기
___
##### 1. 에볼라 데이터프레임을 생성하고 데이터 살펴보기

```python
# 데이터 프레임 생성
ebola = pd.read_csv('/content/drive/MyDrive/esaa/country_timeseries.csv', parse_dates = [0]) # datatime 오브젝트를 인덱스로 설정
```

```python
# 가장 앞쪽 데이터 확인
print(ebola.iloc[:5,:5])
```

```python
# 가장 뒷쪽 데이터 확인
print(ebola.iloc[-5:, :5])
```

##### 2. **date_range** 메서드를 사용하여 시간 인덱스 생성하기
date_range 메서드를 사용하여 2014년 12월 31일부터 2015년 1월 5일의 시간 인덱스(Datetimeindex)를 생성한다.

```python
head_range = pd.date_range(start = '2014-12-31', end = '2015-01-05') 
print(head_range)
```

##### 3. 앞에서 생성한 시간 범위를 인덱스로 지정

```python
# 원본 데이터 손상을 방지하기 위해 앞쪽 데이터만 복사
ebola_5 = ebola.head( )
```

```python
# Date 열을 인덱스로 설정 후, 시간 범위로 인덱스 변경
ebola_5.index = ebola_5['Date']
ebola_5.reindex(head_range) 
print(ebola_5.iloc[:5, :5])
```
