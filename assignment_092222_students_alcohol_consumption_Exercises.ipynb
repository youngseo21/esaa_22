### Step 5. Create a lambda function that will capitalize strings.

```python
def capital():
  return lambda x: x.capitalize()
```

### Step 6. Capitalize both Mjob and Fjob

```python
# DataFrame.capitalize()은 불가능, apply를 사용해 각각 설정해야 된다.
df['Mjob'].apply(capital())
df['Fjob'].apply(capital())
df.head()
```

### Step 8. Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.

### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def majority(i):
  if i > 17:
    return True
  else: 
    return False

df['legal_drinker'] = df.age.apply(lambda age: majority(age))
df.head()
```

```python
def majority(x):
    if x == 1:
        return True
    else:
        return False
df["legal_drinker"] = [majority(1) if x>=18 else majority(0) for x in df["age"]]
df.head()
```
