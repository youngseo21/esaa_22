### | 1) 데이터 시각화가 필요한 이유
---
#####  [ **앤스콤 4분할 그래프(Anscombe's quartet)** 살펴보기 ]
- 앤스콤 4분할 그래프 : 데이터를 시각화하지 않고 수치만 확인할 때 발생할 수 있는 '함정'을 보여주기 위해 만든 그래프
  + 앤스콤 4분할 그래프를 구성하는 데이터 집합은 4개의 그룹으로 구성되어 있으며, 모든 그룹은 x, y 열을 가지고 있다.
  + 이 4개의 구릅은 각각 평균, 분산과 같은 수치값이나 상관관계, 회귀선이 같다
  + 따라서 **'데이터 그룹 1~4의 데이터는 모두 같을 것이다'** 라는 '함정'에 빠질 수 있다.
- 그러나 앤스콤 4분할 그래프를 이용하여 데이터를 시각화하면 각 그룹이 서로 다른 데이터 패턴을 가지고 있다는 것을 쉽게 알 수 있다.

2. 산점도 그래프 : 변수 2개를 사용해서 데이터의 분포를 나타내는 그래프. '이변량 그래프'에 해당한다.

````python
# tips 데이터 집합 호출
# tips 데이터프레임은 지불 금액, 팁, 성별, 흡연 유무, 요일, 시간, 전체 인원 정보를 담고 있다.
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
````

````python
# 히스토그램 그리기

# 기본 틀과 그래프 격자 구성
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
````

````python
# scatter 메서드
axes1.scatter(tips['total_bill'], tips['tip'])

# 그래프와 x축, y축 제목 설정
axes1.set_title('Scatterplot of Total Bill vs. Tips')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

scatter_plot
````

````python
### 박스 그래프 그리기

# 기본 틀과 그래프 격자 구성
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

# boxplot 메서드
axes1.boxplot([tips[tips['sex'] == 'Female']['tip'], tips[tips['sex'] == 'Male']['tip']],
              labels = ['Female', 'Male'])

# 그래프와 x축, y축 제목 설정
axes1.set_xlabel('Sex') 
axes1.set_ylabel('Tip') 
axes1.set_title('Boxplot of Tips by Sex')
````

##### [ 다변량 그래프 그리기 ]
- **다변량 그래프** : 3개 이상의 변수를 사용한 그래프

##### [실습] 다변량 데이터로 다변량 그래프 그리기 - 산점도
  1. 산점도 그래프에서 색상으로 성별을 구분
  2. 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용할 수 없다.
  3. 따라서 0, 1과 같은 정수로 치환해야 한다.
  4. 테이블 당 인원 수(size)를 점의 크기로 구분

````python
# 여성(Female)인 경우 0을 반환하고 남성(Male)인 경우 1을 반환하는 함수
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
````

````python
# recode_sex 메서드가 반환한 값(0, 1)을 데이터프레임에 추가
tips['sex_color'] = tips['sex'].apply(recode_sex)
````

````python
# 기본 틀과 그래프 격자 구성
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)

# scatter 메서드
axes1.scatter(x = tips['total_bill'],
              y = tips['tip'],
              s = tips['size']*10, # s = 점의 크기
              c = tips['sex_color'], # c = 점의 색상
              alpha = 0.5) # alpha = 점의 투명도

# 그래프와 x축, y축 제목 설정
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip')
````

### | 3) seaborn 라이브러리 자유자재로 사용하기
---
##### [실습] 단변량 그래프 그리기 - 히스토그램
  1. Seaborn 라이브러리로 히스토그램 그리기 : `subplots`, `distplot` 메서드 사용
    + `subplots` : 기본 틀 생성
    + `distplot` : 열 데이터를 전달하여 히스토그램 그리기
    + `kde = False` : 밀집도 그래프 제외
    + `hist = False` : 히스토그램 제외
    + `rug = True` : 양탄자 그래프 추가

````python
# tips 데이터 집합 호출
tips = sns.load_dataset('tips')
````

````python
# 기본 틀 생성
ax = plt.subplots()

# 열 데이터를 전달하여 히스토그램 그리기
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot') 
````

````python
# 밀집도 그래프 제외하고 히스토그램 그리기 : kde = False

# 기본 틀 생성
ax = plt.subplots()

# 열 데이터를 전달하여 히스토그램 그리기
ax = sns.distplot(tips['total_bill'], kde = False)

# 제목 설정
ax.set_title('Total Bill Histogram') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Frequency')
````

````python
# 히스토그램 제외하고 밀집도 그래프만 그리기 : hist = False

# 기본 틀 생성
ax = plt.subplots()

# 열 데이터를 전달하여 히스토그램 그리기
ax = sns.distplot(tips['total_bill'], hist = False)

# 제목 설정
ax.set_title('Total Bill Density') 
ax.set_xlabel('Total Bill') 
ax.set_ylabel('Unit Probability')
````

````python
# 양탄자 그래프 추가 : rug = True

# 기본 틀 생성
ax = plt.subplots()

# 열 데이터를 전달하여 히스토그램 그리기
ax = sns.distplot(tips['total_bill'], rug = True)

# 제목 설정
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill') 
````

##### [실습] 다양한 종류의 이변량 그래프 그리기
  1. Seaborn 라이브러리로 산점도 그래프 그리기 : `regplot` 메서드 사용
    + `regplot` : 산점도 그래프와 회귀선을 함께 그릴 수 있다.
    + `fig_reg = False` : 회귀선 제거


````python
# 산점도 그래프와 회귀선 그리기

# 기본 틀 생성
ax = plt.subplots()

# regplot 메서드
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips)

# 제목 설정
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
````

````python
# 회귀선 제거하고 산점도 그래프만 그리기: fit_reg = False

# 기본 틀 생성
ax = plt.subplots( )

# regplot 메서드
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False) 

# 제목 설정
ax.set_title('Scatterplot of Total Bill and Tip') 
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
````

  2. 산점도 그래프와 히스토그램을 그려주는 `jointplot` 메서드
    + x, y 인자에 원하는 열 이름 지정
    + data 인자에는 데이터 프레임 지정
    + 산점도 그래프는 점이 겹쳐 보일 경우 점을 구분하기 어렵다는 단점이 있다.
    + 육각 그래프를 사용하면 산점도그래프의 데이터를 구분하기 쉽게 그릴 수 있다.
      * 육각 그래프 : 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프
      * 특정 데이터의 개수가 많아지면 점점 진한 색으로 표현된다.
      * `kind = hex`


````python
# 산점도 그래프와 히스토그램을 그리기
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)

# x, y축, 그래프 이름 추가
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip') 
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
````

````python
# 육각 그래프 그리기: kind = 'hex
hexbin = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')

# x, y축, 그래프 이름 추가
hexbin.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip',
                    fontsize = 10, y = 1.03)
````

  3. 이차원 밀집도 그리기
    + `kdeplot` 메서드 사용

  7. 관계 그래프 그리기
    + 관계 그래프 : 산점도 그랲, 이차원 밀집도 등 위에서 다룬 그래프를 종합한 그래프
    + `pairplot` 메서드 사용
    + 관계 그래프는 중복된 정보가 표현된다는 단점이 있는데, 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체할 수 있다.
      * `map_upper` 메서드 : 대각선을 기준으로 위쪽에 그릴 그래프 지정
      * `map_lower` 메서드 : 대각선을 기준으로 아래쪽에 그릴 그래프 지정

### | 5) seaborn 라이브러리로 그래프 스타일 설정하기
---
- 지금까지 만든 그래프는 모두 기본 스타일
- seaborn 라이브러리로 만든 그래프의 스타일은 `set_style` 메서드를 사용하여 변경 가능하다.

##### [실습] 그래프에 스타일 적용하기

````python
# 바이올린 그래프 그리기
fig, ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', hue = 'sex', data= tips, split = True)
````

````python
# whitegrid로 스타일 설정 : 그래프의 배경에 가로줄 생성
sns.set_style('whitegrid')
fix, ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', data=tips, hue = 'sex', split = True)
````

````python
# for문을 이용하여 모든 스타일을 하나씩 적용
fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx + 1
  with sns.axes_style(style):
    ax = fig.add_subplot(2,3,plot_position)
    violin = sns.violinplot(x = 'time', y = 'total_bill', data = tips, ax = ax)
    violin.set_title(style)

fig.tight_layout()
````
